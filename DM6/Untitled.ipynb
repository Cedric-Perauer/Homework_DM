```python
import pandas as pd
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
filename_autolabeling_file='out.txt'
global imageDirInput
imageDirInput='/home/cedricperauer/Udacity_video/out_video/'

global imageDirTensorFlow # Folder must exist. Subfolders are created automatically
imageDirTensorFlow='/home/cedricperauer/Udacity_video/street_objects/'

global filename_classnames
filename_classnames='coco.names'
```

```python
# Input Parameter : 1-420
def getFilename(frameID):
    return imageDirInput+"Udacity_Dataset1_frame_{0}.png".format(frameID-1)

# Input Parameter: 1 row of the autolabled dataframe
def getClassID(autolabel): 
    return autolabel['class']

def getFullImage(autolabel):
    frameID=int(autolabel['frame'])
    filename=getFilename(frameID)
    print('filename:'+filename)
    return Image.open(filename)

def getCropImage(fullimage, autolabel):
    width, height = fullimage.size
    left=int((autolabel['x']-0.5*autolabel['w'])*width)
    right=int((autolabel['x']+0.5*autolabel['w'])*width)
    top=int((autolabel['y']-0.5*autolabel['h'])*height)
    bottom=int((autolabel['y']+0.5*autolabel['h'])*height)
    #print(left,top,right,bottom,width,height)
    return fullimage.crop((left,top,right,bottom))

def getClassName(classID):
    df_names=pd.read_csv(filename_classnames,names=['class_name'])
    return df_names.iloc[classID,0]

```

```python
df=pd.read_csv(filename_autolabeling_file,delim_whitespace=True, names=['frame','class','x','y','w','h'])
df.head(20)
```

```python
occuring_classes_list=df['class'].unique()
#occuring_classes_list
occuring_classes_names=getClassName(occuring_classes_list)
occuring_classes_names
```

```python
#occuring_classes_names.iloc[3]
#occuring_classes_names[1]
for row in occuring_classes_names:
    foldername=imageDirTensorFlow+row
    print(foldername)
    os.mkdir(foldername)
```

```python


frameIDOld=0
for index, row in df.iterrows():
    frameID=int(row.frame)
    # Many objects/labels can be in one image. In this case, load image only once
    if frameID>frameIDOld:
        frameIDOld=frameID
        img=getFullImage(row)
    print(frameID)
    classname=getClassName(int(getClassID(row)))
    print(classname)
    cropImage=getCropImage(img,row)
    filename=imageDirTensorFlow+classname+'/'+'image'+str(index)+'.jpg'
    print(filename)
    cropImage.save(filename, "JPEG")

```
